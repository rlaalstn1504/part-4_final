## 1. Docker 기초 개념

### Docker란 무엇인가
Docker는 애플리케이션과 그 실행 환경(라이브러리, 의존성, 설정 등)을 하나의 **컨테이너**라는 단위로 묶어 배포할 수 있게 해주는 플랫폼입니다.  
과거에는 프로그램을 실행하기 위해 OS, 라이브러리, 패키지 버전까지 일일이 맞춰야 했습니다. 하지만 Docker 컨테이너는 "언제 어디서 실행하더라도 동일한 환경"을 보장합니다.  

예시:  
- 마치 여행할 때 필요한 물건을 일일이 가방에 챙기는 대신, 이미 완벽히 준비된 **여행 가방(컨테이너)** 을 가져가서 어디서든 사용할 수 있는 것과 같습니다.

### 시장 수요
- 클라우드 환경에서 일관성 있는 애플리케이션 실행
- 팀 개발 시 "내 컴퓨터에서는 되는데 왜 네 컴퓨터에서는 안 돼?" 문제 해결
- AI/데이터 사이언스 분야에서 복잡한 환경(파이썬 버전, 라이브러리 호환성 등)을 손쉽게 재현 가능

### 컨테이너와 이미지
- **이미지(Image)**: 실행에 필요한 설계도(코드, 라이브러리, 설정이 포함된 패키지)
- **컨테이너(Container)**: 이미지를 기반으로 실행된 실제 프로세스 환경

즉, 이미지는 "레시피", 컨테이너는 "레시피대로 만들어진 음식"이라 할 수 있습니다.

### 컨테이너의 장점
- 어디서든 동일한 실행 환경 제공  
- 빠른 배포 및 확장 가능  
- 마이크로서비스 아키텍처 구현 용이

## 2. Docker 이미지 다루기

### 이미지 관리하기
이미지는 애플리케이션 실행에 필요한 파일(코드, 런타임, 라이브러리, 설정)을 묶어 둔 읽기 전용 템플릿입니다.
컨테이너는 이 이미지를 기반으로 실제 실행되는 프로세스입니다.

#### Docker Hub란?
Docker Hub는 Docker 이미지의 중앙 저장소(레지스트리)입니다.
- 팀이나 개인이 이미지를 업로드(push)하고,
- 다른 개발자가 같은 이미지를 다운로드(pull)해서 같은 환경으로 실행할 수 있습니다.
- `python`, `ubuntu`, `nginx`, `postgres` 같은 공식 이미지도 Docker Hub에서 제공됩니다.

Docker Hub: [https://hub.docker.com/](https://hub.docker.com/)

#### 이미지 사용의 기본 흐름
1. 필요한 이미지를 가져온다 (`pull`)
2. 로컬 이미지 목록을 확인한다 (`ls`)
3. 이미지를 실행해 컨테이너를 만든다 (`run`)
4. (선택) 더 이상 필요 없는 이미지 정리 (`rm`, `prune`)

주요 명령어:
```bash
docker image pull python:3.9    # python:3.9 이미지 다운로드
docker image ls                 # 로컬 이미지 목록 확인
docker container run -it python:3.9 bash  # 이미지로 컨테이너 실행

# 아래 2개는 정리 단계(선택)
docker image rm <IMAGE_ID>      # 특정 이미지 삭제
docker image prune              # 사용하지 않는 이미지 정리
```

#### 2단계 : 도커 이미지를 이용해 컨테이너 실행하기

아래 명령어는 **컨테이너를 새로 생성하면서 즉시 bash 쉘로 진입**합니다.
즉, `run` 한 번으로 "생성 + 시작 + 접속"이 동시에 일어납니다.

```bash
# Linux/macOS 예시
docker run -it --name my_python_container -p 8000:8000 -v "$PWD":/home python:3.9 bash
```

```bash
# Windows PowerShell 예시
docker run -it --name my_python_container -p 8000:8000 -v "${PWD}:/home" python:3.9 bash
```

`$PWD`(PowerShell의 `${PWD}`)는 **현재 터미널 위치한 폴더 경로**를 의미합니다.
그래서 지금 작업 중인 폴더를 컨테이너의 `/home`에 바로 연결할 수 있습니다.

원하면 변수 대신 절대경로를 직접 써도 됩니다.

```bash
# Linux/macOS 절대경로 예시
docker run -it --name my_python_container -p 8000:8000 -v /home/user/project:/home python:3.9 bash
```

```bash
# Windows 절대경로 예시
docker run -it --name my_python_container -p 8000:8000 -v C:/Users/username/project:/home python:3.9 bash
```

이미 실행 중인 컨테이너에 다시 들어갈 때는 `exec`를 사용합니다.

```bash
docker exec -it my_python_container bash
```

추가로 알아두면 좋은 흐름:
- 쉘에서 `exit`를 입력하면 컨테이너의 메인 프로세스가 종료되어 컨테이너도 멈출 수 있습니다.
- 컨테이너를 종료하지 않고 빠져나오려면 `Ctrl+P` 후 `Ctrl+Q`를 사용합니다.

GPU가 있는 환경에서만 `--gpus all`을 추가하세요.

| 옵션 | 설명 | 예시 |
| --- | --- | --- |
| `-i` | 표준 입력을 열어 둠 | `-i` |
| `-t` | 터미널(TTY) 할당 | `-t` |
| `-it` | `-i` + `-t` 조합 | `-it` |
| `--name` | 컨테이너 이름 지정 | `--name my-container` |
| `-p` | 호스트:컨테이너 포트 매핑 | `-p 8000:8000` |
| `--gpus all` | GPU 장치 전달(선택) | `--gpus all` |
| `-v` | 호스트 경로를 컨테이너 경로에 마운트 | `-v "$PWD":/home` |
| `bash` | 컨테이너 시작 후 실행할 명령 | `python:3.9 bash` |

참고: 실습용 쉘 컨테이너에는 보통 `--restart=always`를 사용하지 않습니다.

### Dockerfile이란?
`Dockerfile`은 이미지를 만드는 과정을 코드로 적어둔 파일입니다.
한 줄씩 위에서 아래로 실행되며, 각 줄은 이미지의 레이어(layer)를 만듭니다.

Dockerfile을 사용하면 다음이 쉬워집니다.
- 같은 환경을 반복해서 만들기
- 팀원이 같은 방식으로 이미지 빌드하기
- 변경 이력을 Git으로 관리하기

### Dockerfile의 주요 구성 요소
1. **FROM**: 시작점이 되는 베이스 이미지를 선택
2. **WORKDIR**: 이후 명령이 실행될 기본 디렉토리 지정
3. **COPY**: 로컬 파일을 이미지 내부로 복사
4. **RUN**: 패키지 설치/설정 명령 실행
5. **CMD** 또는 **ENTRYPOINT**: 컨테이너 시작 시 실행할 기본 명령 지정
    - CMD: 기본 명령을 정하며, 실행 시 쉽게 변경 가능. 사용자가 docker run에서 다른 명령을 주면 쉽게 대체
    - ENTRYPOINT: 항상 실행될 주 명령을 고정하며, 컨테이너를 전용 실행 파일처럼 만들기 좋음
    - CMD는 유연한 기본값, ENTRYPOINT는 고정된 실행 중심점

### Dockerfile 작성하기

여기서는 **`requirements.txt`가 없는 프로젝트**를 기준으로 설명합니다.

기본 흐름:
1. 어떤 기반 이미지에서 시작할지 정한다 (`FROM`)
2. 작업 폴더를 정한다 (`WORKDIR`)
3. 애플리케이션 코드를 복사한다 (`COPY . .`)
4. 실행 명령을 지정한다 (`CMD`)

프로젝트에 외부 라이브러리가 있다면,
- `requirements.txt`를 만든 뒤 설치 단계를 추가하거나
- `RUN pip install 패키지명` 형태로 필요한 패키지를 직접 설치하면 됩니다.

그리고 Dockerfile의 큰 장점은 **매우 작은 텍스트 파일**이라는 점입니다.
완성된 Docker 이미지는 수백 MB~수 GB가 될 수 있지만, Dockerfile은 보통 몇 KB 수준입니다.
그래서 이미지 파일을 직접 전달하기보다,
- Dockerfile + 소스코드를 공유하고
- 각 환경에서 동일 절차로 이미지를 빌드하는 방식이 협업/배포에 훨씬 유리합니다.


아래는 Python 기반 앱을 위한 **Dockerfile 예시**입니다.

프로젝트에 `requirements.txt`가 **없다**고 가정한 기본 예시:

```dockerfile
FROM python:3.11-slim

# 컨테이너 내부 작업 폴더
WORKDIR /app

# 애플리케이션 코드 복사
COPY . .

# 컨테이너 시작 시 실행할 기본 명령
CMD ["python", "app.py"]
```

외부 패키지가 필요한 경우에는 두 가지 방법 중 하나를 사용하면 됩니다.

1) `requirements.txt`를 만들어 설치
```dockerfile
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
```

2) 필요한 패키지를 직접 설치
```dockerfile
RUN pip install --no-cache-dir fastapi uvicorn
```

---

아래처럼 Dockerfile에서 GitHub 저장소를 직접 받아 빌드하는 방식도 사용할 수 있습니다.

```dockerfile
FROM python:3.11-slim

WORKDIR /app
RUN apt-get update && apt-get install -y git
RUN git clone https://github.com/rlaalstn1504/sample_2.git .
RUN pip install --no-cache-dir -r requirements.txt

CMD ["python", "app.py"]
```

이 방식은 "저장소 기준으로 즉시 빌드"할 때 편리합니다.
다만 위 예시는 저장소 안에 `requirements.txt`가 있다고 가정합니다.


#### 설명

- **FROM**
  - 어떤 운영 환경에서 시작할지 정합니다.

- **WORKDIR**
  - 이후 명령이 실행될 기본 작업 폴더를 정합니다.

- **COPY . .**
  - 현재 프로젝트 파일 전체를 컨테이너로 복사합니다.

- **CMD**
  - 컨테이너 시작 시 기본으로 실행할 명령입니다.

추가 설치가 필요할 때:
- `RUN pip install ...`로 패키지를 직접 설치하거나
- `requirements.txt`를 만든 뒤 `RUN pip install -r requirements.txt`를 사용하면 됩니다.

확인 포인트:
- 빌드는 성공했는데 실행 시 실패하면, `CMD`의 파일 경로(`app.py`)와 `WORKDIR`가 맞는지 먼저 점검하세요.

```bash
docker container run --rm -it my-python-app:1.0
```

이미지를 로컬에서 만든 후 Docker Hub에 업로드하면 다른 환경에서 쉽게 재사용할 수 있습니다.

### 2. Docker Hub에 배포하기
1. 로그인
```bash
docker login
```

2. 태그 지정
```bash
docker image tag my-python-app:1.0 <username>/my-python-app:1.0
```
- `<username>`: Docker Hub 계정명(네임스페이스)
- `my-python-app`: 저장소 이름
- `1.0`: 버전 태그

3. 푸시
```bash
docker image push <username>/my-python-app:1.0
```

태그 전략 예시:
- 릴리스 버전: `1.0.0`, `1.1.0`
- 날짜 버전: `2026-02-25`
- 테스트용 별도 태그: `dev`, `staging`


#### Dockerfile과 Docker Hub의 차이점 및 역할

- **Docker Hub**  
  - 공용 이미지 저장소(마트의 기본 재료).  
  - python, ubuntu, nginx 등 이미 완성된 기본 환경을 가져올 때 사용.  
  - `docker pull python:3.9` 같은 명령으로 다운로드.

- **Dockerfile**  
  - 나만의 이미지를 만드는 **레시피(설계도)**.  
  - 기본 이미지 위에 내 애플리케이션, 라이브러리, 설정 등을 추가하여  
    **커스텀 이미지를 자동으로 생성**할 수 있음.  
  - 파일 크기가 매우 작아 팀 개발·배포에 유리함.  
  - “환경을 복사해서 보내는 것이 아니라,  
    **환경을 만드는 방법을 코드로 공유**하는 것”이 Dockerfile의 핵심 가치.

### Dockerfile을 사용하는 가장 큰 장점

1. **완전한 재현성**  
   - Dockerfile은 실행 환경을 코드 형태로 정의하기 때문에  
     어디서 빌드하든 동일한 이미지가 만들어집니다.  
   - GitHub에서 소스를 클론하는 방식이라면 로컬 파일이 없어도 동일한 환경을 재생성할 수 있습니다.

2. **협업 효율성 향상**  
   - 팀원 모두가 같은 Dockerfile을 사용하면 동일한 개발 환경을 유지할 수 있어  
     “환경이 달라서 안 돌아요” 같은 문제가 사라집니다.

3. **배포 용이성과 가벼운 공유**  
   - Docker 이미지 자체는 수백 MB~수 GB로 매우 크지만,  
     Dockerfile과 소스코드는 매우 가볍습니다.  
   - 이미지 파일을 직접 전달할 필요 없이,  
     *Dockerfile + GitHub repository*만 공유하면 누구나 동일 이미지 생성 가능.

4. **버전 관리의 이점**  
   - Dockerfile도 소스코드와 함께 Git으로 버전 관리할 수 있어  
     변경 이력 추적, 협업, 롤백이 매우 용이합니다.

#### 핵심 요약

- Docker Hub = 공용 기본 이미지 저장소  
- Dockerfile = 나만의 개발·실행 환경을 만드는 레시피  
- Dockerfile은 이미지보다 훨씬 가벼워서 배포·공유에 적합  
- Dockerfile을 사용하면 어떤 PC에서도 같은 환경을 자동으로 재생성할 수 있음



### 이미지 빌드 및 배포

이미지를 로컬에서 만든 후, 이를 Docker Hub와 같은 외부 저장소에 배포하여 다른 사람들과 공유할 수 있습니다.

### 1. 이미지 빌드하기
`docker build` 명령어를 사용하여 `Dockerfile`을 기반으로 이미지를 빌드합니다:
```bash
docker image build -t my-python-app .
```
여기서 `-t my-python-app`은 빌드된 이미지를 `my-python-app`이라는 태그로 지정하는 부분입니다.

### 2. Docker Hub에 배포하기
이미지를 Docker Hub에 배포하려면 먼저 Docker Hub에 로그인한 후, 이미지를 푸시할 수 있습니다.

1. **Docker Hub에 로그인**:
   ```bash
   docker login
   ```

2. **이미지에 태그 지정**:
   ```bash
   docker image tag my-python-app <username>/my-python-app:latest
   ```
   여기서 `<username>`은 Docker Hub 계정 이름입니다.

3. **이미지 푸시**:
   ```bash
   docker image push <username>/my-python-app:latest
   ```

위 과정을 통해 로컬에서 만든 이미지를 Docker Hub에 업로드하고, 이를 다른 서버나 팀원들이 사용할 수 있도록 공유할 수 있습니다.

## 3. Docker 컨테이너 실행 및 관리

### 컨테이너 실행하기
컨테이너는 Docker 이미지를 기반으로 실행되는 **실제 인스턴스**입니다.  
`docker container run` 명령어를 사용하여 새로운 컨테이너를 실행할 수 있습니다. 

### 기본 명령어
```bash
docker container run -d --name myapp -p 5000:5000 my-python-app
```
- `-d`: **백그라운드 실행**을 의미합니다. 즉, 터미널에서 컨테이너를 실행하고 나서도 해당 터미널을 계속 사용할 수 있습니다.
- `--name myapp`: 실행되는 컨테이너에 **`myapp`** 이라는 이름을 지정합니다. 이를 통해 나중에 컨테이너를 쉽게 관리할 수 있습니다.
- `-p 5000:5000`: **포트 매핑**을 설정합니다. 이는 호스트 머신의 포트 5000을 컨테이너의 포트 5000에 연결하는 방식입니다.
  - 포트 매핑: 컨테이너는 기본적으로 외부에서 바로 접근 불가능한 독립된 네트워크에 있습니다. 포트 매핑은 호스트 머신과 컨테이너 사이에 네트워크 요청이 전달될 수 있는 경로를 설정하는 기능입니다.
  - 예를 들어, 웹 애플리케이션을 실행하는 컨테이너가 있을 경우, 호스트 머신에서 해당 애플리케이션에 접근하려면 **호스트 머신의 포트**가 컨테이너의 포트에 매핑되어야 합니다.

#### 포트란 무엇인가?
- **포트(Port)** 는 네트워크에서 **특정 서비스나 애플리케이션**을 식별하는 번호입니다. 이를 통해 클라이언트는 서버의 특정 애플리케이션에 접근할 수 있습니다.
- 예를 들어, 웹 서버는 일반적으로 **80번 포트**를 사용하고, 보안 웹 서버는 **443번 포트**를 사용합니다.
- Docker에서는 **호스트 머신의 포트**와 **컨테이너의 포트**를 매핑하여, 외부에서 컨테이너 내부의 서비스를 사용할 수 있도록 합니다.

### 컨테이너 관리

컨테이너를 실행한 후, 이를 관리하고 모니터링하기 위한 명령어들이 있습니다.

### 주요 명령어:
- `docker container ls`: 현재 실행 중인 컨테이너 목록을 보여줍니다.
- `docker container exec -it myapp bash`: 실행 중인 `myapp` 컨테이너에 **터미널을 접속**하여 명령어를 실행할 수 있습니다. `-it` 옵션은 **인터랙티브** 모드로 터미널을 실행합니다.
- `docker container logs myapp`: 컨테이너의 로그를 확인합니다. 예를 들어, `myapp` 컨테이너의 로그를 출력하여 오류나 실행 상태를 모니터링할 수 있습니다.
- `docker container stop myapp`: `myapp` 컨테이너를 **정지**시킵니다.
- `docker container rm myapp`: 정지된 컨테이너를 **삭제**합니다. 컨테이너가 정지 상태여야만 삭제할 수 있습니다.
- `docker container prune`: **사용하지 않는 모든 컨테이너**를 한 번에 삭제합니다.

#### Docker 네트워크란?

컨테이너들이 서로 통신할 수 있도록 연결하는 **가상 네트워크 영역**입니다.  
기본 bridge 네트워크만으로도 통신은 가능하지만, 컨테이너 IP(사설IP)는 재시작 시 바뀔 수 있어 관리가 어렵습니다.  
사용자 정의 네트워크를 사용하면 IP 변화에 상관없이 **컨테이너 이름(DNS)** 으로 안정적인 통신이 가능합니다.


#### 왜 사용자 정의 네트워크를 사용할까?

- **컨테이너 IP 변경 문제 해결**  
  IP가 변경되어도 컨테이너 이름으로 접속할 수 있어 환경이 안정적임.

- **DNS 기반 이름 통신 지원**  
  같은 네트워크 내 컨테이너는 `web -> db:3306`처럼 이름만으로 접근 가능.

- **보안 및 서비스 분리**  
  서로 관련 없는 서비스는 다른 네트워크에 배치해 통신을 제한할 수 있음.

- **멀티컨테이너 애플리케이션 구성에 필수**  
  웹, DB, 캐시 등 여러 컨테이너를 하나의 시스템처럼 묶어 관리하기 쉬움.


#### 네트워크 생성 및 컨테이너 연결

    docker network ls
    docker network create mynetwork
    docker container run -d --network mynetwork my-python-app

- `docker network create mynetwork`  
  컨테이너를 묶기 위한 사용자 정의 네트워크 생성.

- `docker container run -d --network mynetwork my-python-app`  
  컨테이너를 특정 네트워크에 연결해 실행.



#### IP 고정 vs 컨테이너 이름 사용 (추가)

컨테이너는 실행 시 내부 IP를 할당받으며, 재시작하면 해당 IP는 변경될 수 있습니다.  
필요한 경우 `--ip` 옵션을 사용해 IP를 직접 고정할 수 있으나,  
IP 충돌 위험과 확장·관리의 어려움으로 인해 일반적으로 권장되지는 않습니다.

따라서 Docker 환경에서는 IP에 의존하기보다,  
**사용자 정의 네트워크에서 제공되는 컨테이너 이름(DNS)을 사용한 통신 방식이 권장됩니다.**

예:
- (IP 고정 방식, 비권장) `172.18.0.10:3306`
- (이름 기반 방식, 권장) `db:3306`

#### Docker 데이터 저장(마운트)

컨테이너 내부 파일 시스템은 임시 저장소이기 때문에, 컨테이너가 삭제되면 내부 데이터도 함께 사라집니다.  
지속적으로 유지해야 하는 데이터는 **마운트(Mount)** 기능을 사용하여 컨테이너 외부에 보존할 수 있습니다.

#### 바인드 마운트(Bind Mount)

바인드 마운트는 **호스트의 실제 디렉토리**를 컨테이너 내부 경로에 그대로 연결하는 방식입니다.

    docker run -d -v /host/path:/container/path my-app

- `/host/path` : 호스트 시스템의 실제 디렉토리 경로  
- `/container/path` : 컨테이너 내부에서 해당 데이터가 보이는 위치  
- 컨테이너가 종료되거나 삭제되더라도 `/host/path`에 저장된 데이터는 그대로 유지됩니다.  
- 개발 환경에서 소스 코드나 설정 파일을 실시간으로 반영할 때 가장 많이 사용되는 방식입니다.

### 4.1 Docker Compose 소개
Docker Compose는 여러 컨테이너를 하나의 YAML 파일(`docker-compose.yml`)로 정의하고,
하나의 프로젝트처럼 함께 실행/중지/로그확인을 할 수 있게 해주는 도구입니다.

예를 들어 웹 서비스는 보통 단일 컨테이너로 끝나지 않습니다.
- 프론트엔드(Nginx)
- 백엔드 API(FastAPI)
- 데이터베이스(PostgreSQL)

Compose를 쓰면 위 구성을 한 파일에서 관리할 수 있어,
실행 명령이 단순해지고 환경 재현이 쉬워집니다.

가장 자주 사용하는 명령:
- `docker compose up -d` : 서비스 전체 시작 (-d 없이 실행: 터미널 창에 모든 컨테이너의 로그가 실시간으로 출력)
- `docker compose down` : 서비스 전체 중지(기본적으로 볼륨은 유지)

참고: 예전 표기인 `docker-compose`(하이픈)도 여전히 보이지만,
현재 Docker Desktop/CLI에서는 `docker compose`(공백) 사용을 권장합니다.


#### `docker-compose.yml` 예시

아래는 **정적 웹(Nginx) + API(FastAPI) + DB(PostgreSQL)** 로 구성된 수업용 `docker-compose.yml` 예시입니다.  
서비스 이름을 기준으로 컨테이너 간 내부 통신을 합니다.

```yaml
services:
  # 1) 정적 웹 프론트엔드 (Nginx)
  # - 브라우저는 http://localhost:8080 로 접속
  # - /api/* 요청은 내부 API 컨테이너로 프록시됨
  web:
    image: nginx:alpine            # 가벼운 Nginx 이미지 사용
    ports:
      - "8080:80"                  # 호스트 8080 → 컨테이너 80
    volumes:
      - ./web:/usr/share/nginx/html:ro                 # 정적 파일(HTML/CSS/JS)
      - ./web/nginx.conf:/etc/nginx/conf.d/default.conf:ro  # 프록시 설정
    depends_on:
      - api                         # API가 먼저 떠야 웹이 정상 동작

  # 2) 간단한 API (FastAPI)
  # - DB에서 rows를 읽어 JSON으로 반환
  # - web 컨테이너는 http://api:8000 로 통신
  api:
    build: ./api                   # ./api/Dockerfile 빌드
    environment:
      DATABASE_URL: postgresql://app:app@db:5432/appdb  # 서비스명 db로 접속
    ports:
      - "8000:8000"                # (선택) 호스트에서 직접 API 확인
    depends_on:
      - db                          # DB가 먼저 떠야 API가 연결 가능

  # 3) 데이터베이스 (PostgreSQL)
  # - init.sql로 샘플 테이블/데이터 자동 생성
  db:
    image: postgres:16-alpine      # 가벼운 Postgres 이미지
    environment:
      POSTGRES_USER: app           # DB 사용자
      POSTGRES_PASSWORD: app       # DB 비밀번호
      POSTGRES_DB: appdb           # 기본 DB 이름
    volumes:
      # - db_data:/var/lib/postgresql/data             # (수업용) 데이터 영속화가 필요 없으면
      - ./db/init.sql:/docker-entrypoint-initdb.d/init.sql:ro  # 초기 스키마/데이터

# (수업용) 영속 볼륨은 사용하지 않음
# volumes:
#   db_data:

```

**`services:`** 애플리케이션을 구성하는 컨테이너 묶음  
**`web:`** Nginx 기반 정적 웹 서버. 브라우저는 `localhost:8080`으로 접속  
**`api:`** FastAPI 백엔드. DB 접근 시 서비스 이름 `db`를 호스트처럼 사용  
**`db:`** PostgreSQL 데이터베이스. `init.sql`로 초기 테이블/데이터 구성  
- Compose 내부 통신은 IP 대신 서비스 이름(`api`, `db`)을 사용  
- `ports`는 "호스트 포트:컨테이너 포트" 형식  
- 수업 예제는 구조 이해가 목적이라 DB 영속 볼륨을 생략  

### 4.2 Docker Compose 명령어

#### 서비스 실행
`docker compose up` : 현재 디렉토리의 compose 파일 기준으로 서비스를 실행합니다.  
`docker compose up -d` : 백그라운드(detached)로 실행합니다.  
`docker compose up --build` : 실행 전에 이미지를 다시 빌드합니다.  

#### 서비스 중지 및 정리
`docker compose down` : 현재 compose 프로젝트의 컨테이너/네트워크를 중지 및 정리합니다.  
- 기본 동작에서는 named volume이 삭제되지 않습니다.  
- 볼륨까지 삭제하려면 `docker compose down --volumes`를 사용합니다.  

#### 상태/로그 확인
`docker compose ps` : 실행 상태 확인  
`docker compose logs` : 전체 로그 확인  
`docker compose logs -f` : 실시간 로그 추적  

학습 흐름 추천:
1. `up -d`로 실행
2. `ps`로 상태 확인
3. `logs -f`로 동작 확인
4. `down`으로 정리


### 4.3 Docker Compose의 장점
- **설정 일원화**: 여러 컨테이너 설정을 하나의 파일에서 관리
- **실행 단순화**: 한두 개 명령으로 전체 스택 제어
- **협업 효율**: 팀원이 동일한 compose 파일로 같은 환경 재현 가능
- **학습/개발 속도 향상**: 웹/API/DB 연결 구조를 빠르게 검증 가능

주의할 점:
- `depends_on`은 "시작 순서"를 맞춰줄 뿐, 서비스가 "요청 처리 가능한 상태(ready)"인지까지는 보장하지 않습니다.
- API가 DB 준비 전에 뜰 수 있으므로, 실무에서는 healthcheck/재시도 로직을 함께 사용합니다.

## 5. Kubernetes (K8s) 개념

### 5.1 Kubernetes란 무엇인가? (Docker Compose와의 차이점)
위에서 배운 **Docker Compose**는 '한 대의 컴퓨터'에서 여러 개의 컨테이너를 실행하고 묶어주는 훌륭한 도구입니다.  
하지만 서비스가 커져서 접속자가 몰리거나, 그 한 대의 컴퓨터가 통째로 고장 나면 어떻게 될까요?

**Kubernetes(K8s)** 는 이 문제를 해결합니다.  
한 대가 아닌 **여러 대의 컴퓨터(서버 클러스터)** 를 마치 하나의 거대한 컴퓨터처럼 묶어서,  
수많은 컨테이너를 **자동으로 배치, 확장, 유지 보수(오케스트레이션)** 해주는 도구입니다.  
구글이 개발하여 오픈소스로 공개했으며, 현재 클라우드 환경의 핵심 표준입니다.

* **Docker Compose**: 동네 작은 식당의 지배인 (단일 서버, 소규모 수동 관리)
* **Kubernetes**: 전국 프랜차이즈 식당의 중앙 AI 통제 시스템 (다중 서버, 대규모 자동 관리)


참고 : 
- **컨테이너 오케스트레이션**: 여러 개의 컨테이너가 서로 협력하여 하나의 애플리케이션을 이루는 환경에서 이들을 효율적으로 관리하고 조정하는 기술입니다.  
Kubernetes는 이러한 작업을 자동화하여 애플리케이션의 안정성, 확장성, 가용성을 높여줍니다.

- **구글이 개발**하고, 현재는 **CNCF(Cloud Native Computing Foundation)** 에서 관리하고 있습니다.  
Kubernetes는 클라우드 환경에서 **컨테이너화된 애플리케이션**을 운영하는 데 있어 업계 표준 도구로 자리 잡았습니다.

### 5.2 Kubernetes의 핵심 기능 (왜 K8s를 써야 할까?)
Docker Compose만으로는 대응하기 힘든, K8s만의 강력한 자동화 기능들입니다.

**1. 자기 치유 (Self-healing) - "서버가 죽으면 알아서 다시 살려냅니다"**
* **이전의 문제:** 서버 1대가 갑자기 고장 나면, 개발자가 새벽에 자다 깨서 서버를 고치고 컨테이너를 다시 실행해야 했습니다.
* **K8s의 해결:** 특정 컴퓨터의 전원이 나가거나 앱이 뻗어버리면, K8s는 이를 즉시 감지합니다.  
그리고 죽어버린 컨테이너들을 **다른 정상적인 컴퓨터로 자동으로 옮겨서 즉시 다시 실행**시킵니다. 시스템이 알아서 응급처치를 하는 것입니다.

**2. 자동 확장 (Auto-scaling) - "바빠지면 알바생을 자동으로 늘립니다"**
* **이전의 문제:** 트래픽이 몰리면 사람이 직접 서버에 접속해 컨테이너 개수를 늘려야 했습니다.
* **K8s의 해결:** 유명 방송에 우리 웹사이트가 소개되어 접속자가 100배 폭주한다고 가정해 봅시다.  
Kubernetes는 Horizontal Pod Autoscaler(HPA) 를 통해 CPU/메모리 사용률을 기준으로 Pod(컨테이너) 개수를 자동으로 증가/감소시킵니다.  

  단, 이는 현재 클러스터에 남아 있는 자원 범위 내에서의 확장입니다.  
  만약 Pod를 배치할 자원이 부족하면, 클라우드 환경에서는 Cluster Autoscaler가 새로운 노드(서버)를 자동으로 추가하고, 트래픽이 줄어들면 노드도 함께 축소됩니다.  

**3. 무중단 배포 (Rolling Update) - "영업 정지 없이 간판을 바꿉니다"**
* 새로운 버전의 프로그램을 배포할 때, K8s는 기존 서비스를 멈추지 않습니다. 새 버전의 컨테이너를 하나씩 띄우고,  
구 버전 컨테이너를 하나씩 내리는 방식으로 순차 교체합니다. 사용자는 점검 시간 없이 365일 서비스를 이용할 수 있습니다.

**4. 트래픽 분산 (Load Balancing) - "알아서 주문을 나눠줍니다"**
* 컨테이너가 100개로 늘어나면 트래픽을 골고루 분산시켜줘야 합니다. K8s는 늘어난 컨테이너들에게 사용자들의 접속 요청을 알아서 공평하게 나눠줍니다.

### 5.3 언제 Kubernetes를 써야 할까? (사용 사례)

1. **AI/ML 모델 서빙 자동화** 
- > 평소에는 최소한의 자원만 유지하다가, 무거운 AI 이미지 생성/추론 요청이 몰릴 때만 GPU와 컨테이너를 쫙 늘려서 빠르게 처리하고 싶을 때.
2. **대규모 트래픽이 예상되는 웹 서비스** 
- > 수강신청, 명절 기차표 예매, 쇼핑몰 특가 이벤트 등 특정 시간에 트래픽이 폭주해도 서버가 죽지 않고 유연하게 버텨야 할 때.
3. **절대 멈추면 안 되는 금융/결제 시스템**
- > 물리적인 서버 한두 대가 고장 나더라도(정전, 하드웨어 불량 등) 시스템 전체는 1초도 멈추지 않고 '자기 치유'를 통해 계속 운영되어야 할 때.

### 5.4 Kubernetes의 장점
- **자동화된 관리**: 배포, 확장, 복구가 자동으로 이루어져 개발자는 시스템 관리에서 벗어나 애플리케이션 개발에 집중할 수 있습니다.
- **높은 확장성**: Kubernetes는 트래픽에 따라 컨테이너 수를 자동으로 늘리거나 줄여 시스템 리소스를 효율적으로 사용할 수 있습니다.
- **고가용성**: 장애 발생 시 자동으로 복구하고, 서비스의 중단 없이 운영할 수 있습니다.
- **자원 최적화**: 여러 애플리케이션을 하나의 클러스터에서 관리하여 리소스를 효율적으로 사용합니다.


### 5.5 요약정리
* **Docker Compose**가 **내 컴퓨터에서 여러 컨테이너를 띄우고 테스트하는 법**이라면, 
* **Kubernetes**는 **여러 대의 컴퓨터를 엮어서, 컨테이너들이 죽으면 살려내고 바쁘면 늘려가며 24시간 무인으로 돌아가게 만드는 마스터 시스템**입니다.

### Kubernetes 실습

#### 1. 실습 목표

이 실습에서는 Kubernetes의 목적을 명확히 이해하기 위해,
서로 다른 3개의 컨테이너가 연결되어 동작하는 작은 애플리케이션을 배포합니다.

구성 요소는 다음과 같습니다.

1) time-api : 현재 시간을 반환하는 FastAPI 서비스  
2) logger-api : time-api가 전달한 데이터를 출력(log)하는 FastAPI 서비스  
3) web-frontend : time-api를 호출하여 브라우저에 시간 표시하는 간단한 프론트엔드

Kubernetes에서는 각 서비스를 Deployment와 Service로 배포하며,  
서비스 이름을 이용하여 서로 통신할 수 있음을 확인합니다.

#### 실습 환경에 대한 중요한 안내

이 실습은 다음 두 가지 환경을 모두 고려합니다.

1) Windows + WSL에서 Minikube 실행  
2) 다른 PC에 SSH로 접속하여 Minikube 실행 (동일 네트워크)

#### time-api 코드 (요약)
```python
# k8s/time-api/main.py
from datetime import datetime, timezone          # UTC 기준 현재 시간 생성
import os                                       # 환경변수 접근
import requests                                 # logger-api로 HTTP 요청
from fastapi import FastAPI                     # API 서버 프레임워크

app = FastAPI()                                 # FastAPI 앱 생성
LOGGER_URL = os.getenv("LOGGER_URL", "http://logger-api:8000/log")  # logger-api 서비스 주소

@app.get("/time")                               # /time 엔드포인트
def get_time():
    now = datetime.now(timezone.utc).isoformat()  # 현재 UTC 시간을 ISO 문자열로 변환
    try:
        requests.post(LOGGER_URL, json={"iso": now}, timeout=0.5)  # 시간 정보를 logger-api로 전송
    except Exception:
        pass                                    # logger 실패 시에도 time-api 응답은 유지
    return {"iso": now}                         # 클라이언트에 현재 시간 반환
```


#### logger-api 코드 (요약)
```python
# k8s/logger-api/main.py
from fastapi import FastAPI                     # API 서버 프레임워크

app = FastAPI()                                 # FastAPI 앱 생성

@app.post("/log")                               # /log 엔드포인트
def log_time(payload: dict):
    print("[logger-api] payload:", payload)     # 전달받은 데이터 로그 출력
    return {"status": "ok"}                     # 정상 처리 응답
```


#### web-frontend 코드 (요약)
```html
<!-- k8s/web-frontend/index.html -->
<h1>현재 시간</h1>
<div id="time">로딩 중...</div> <!-- API 응답 전 표시 -->
```

#### 11. Minikube 안에서 이미지 빌드

Minikube는 **자체 Docker 환경**을 사용합니다.  
쿠버네티스가 이미지를 바로 사용하려면 **Minikube가 사용하는 Docker에 이미지가 존재해야 합니다.**

**방법 A: 내 터미널을 Minikube Docker로 연결해서 빌드**
```bash
# 현재 터미널을 Minikube가 사용하는 Docker 환경으로 전환
eval $(minikube -p minikube docker-env)

# time-api 이미지를 Minikube Docker 환경에 빌드
docker build -t time-api:1.0 ./k8s/time-api

# logger-api 이미지를 Minikube Docker 환경에 빌드
docker build -t logger-api:1.0 ./k8s/logger-api

# web-frontend 이미지를 Minikube Docker 환경에 빌드
docker build -t web-frontend:1.0 ./k8s/web-frontend
```
- 이 터미널에서 실행되는 `docker build`는 Minikube 내부 Docker를 사용합니다.
- 새 터미널을 열면 다시 `docker-env` 설정이 필요합니다.


**방법 B: `minikube image build` 사용**
```bash
# time-api 이미지를 Minikube 내부 Docker에 빌드
minikube image build -t time-api:1.0 ./k8s/time-api

# logger-api 이미지를 Minikube 내부 Docker에 빌드
minikube image build -t logger-api:1.0 ./k8s/logger-api

# web-frontend 이미지를 Minikube 내부 Docker에 빌드
minikube image build -t web-frontend:1.0 ./k8s/web-frontend
```
- Minikube 내부 Docker에 직접 이미지가 빌드됩니다.
- 별도의 Docker 환경 전환이 필요 없습니다.

#### 12. K8s 핵심 개념 이해 및 매니페스트 적용

명령어를 실행하기 전에, 우리가 무엇을 만들고 있는지 핵심 개념 2가지를 먼저 알아야 합니다.

**개념 1: 매니페스트(Manifest)란?**
매니페스트는 쿠버네티스에게 "이 컨테이너를 이렇게 만들어서 띄워줘"라고 알려주는 **설정 파일(실행 설명서)** 입니다.  
보통 YAML 형식으로 작성되며, 어떤 이미지를 쓸지, 몇 개를 띄울지 등의 정보가 담겨 있습니다.

* `k8s/manifests/` 폴더 안에는 각 앱을 띄우는 설정(Deployment)과 통신을 위한 설정(Service)이 적힌 3개의 파일이 준비되어 있습니다.
  - `time-api.yaml` : time-api를 띄우는 설정 (Deployment) + 내부 통신용 서비스 (Service)
  - `logger-api.yaml` : logger-api를 띄우는 설정 + 내부 통신용 서비스
  - `web-frontend.yaml` : 웹을 띄우는 설정 + 외부에서 접속할 수 있게 하는 서비스

**개념 2: 파드(Pod)란?**
파드는 **쿠버네티스에서 실행되는 가장 작은 단위**입니다. 쿠버네티스는 컨테이너를 맨몸으로 실행하지 않고, 항상 '파드'라는 캡슐로 감싸서 실행합니다.  
우리가 만든 웹 서버, API 서버는 모두 이 파드 안에서 실행됩니다.
  - 컨테이너 하나 또는 여러 개를 포함할 수 있음
  - 우리가 만든 웹 서버, API 서버는 모두 **파드 안에서 실행됨**
  - 컨테이너를 직접 실행하는 것이 아니라,  **항상 파드를 통해 실행**됩니다.

##### 실제 적용 명령 (필수)

이제 빌드한 이미지를 바탕으로, 매니페스트(설명서)를 클러스터에 적용해 보겠습니다.  

이미지를 빌드한 뒤에는 아래 명령으로 매니페스트를 **클러스터에 적용**해야 합니다.  
이 단계를 해야 Deployment/Service가 실제로 생성되며, 이후 `minikube service web-frontend --url`이 동작합니다.

```bash
# 방법 1: manifests 폴더 전체를 한 번에 적용
kubectl apply -f k8s/manifests/

# 방법 2: 파일별로 적용
kubectl apply -f k8s/manifests/time-api.yaml
kubectl apply -f k8s/manifests/logger-api.yaml
kubectl apply -f k8s/manifests/web-frontend.yaml
```

적용 확인:
```bash
# 1. 파드(앱)가 잘 떴는지 확인 (STATUS가 Running인지 확인)
kubectl get pods

# 2. 서비스(네트워크)가 잘 만들어졌는지 확인
kubectl get svc
```

참고: `SVC_NOT_FOUND` 에러는 대부분 이 apply 단계가 누락되었을 때 발생합니다.

#### 13. 웹 접속 방식

Minikube를 Docker 기반으로 실행할 경우, 실습 환경(WSL 또는 SSH 원격 서버)에 따라 접속 환경이 호스트 PC와 분리되어 있습니다.  
따라서 방식 2 (port-forward)를 사용하는 것을 강력히 권장합니다.

##### 방식 1. NodePort 기반 접속 (참고용)

NodePort는 **Kubernetes가 실행 중인 서버(Node)의 포트를 외부에 열어**,  
해당 포트로 들어온 요청을 Service로 전달하는 방식입니다.

```bash
# web-frontend 서비스의 NodePort 접속 주소를 출력
minikube service web-frontend --url
```

명령 실행 시 Minikube가 접근 가능한 URL을 출력합니다.
예: `http://192.168.49.2:30080`

[주의] 이 주소는 Minikube 내부 가상 IP입니다.  
WSL 환경이나 SSH 원격 접속 환경에서는 여러분의 웹 브라우저에서 이 주소로 직접 접속되지 않습니다.  
네트워크가 어떻게 구성되어 있는지 확인하는 용도로만 참고하시고, 실제 접속은 아래 방식 2를 사용하세요.

##### 방식 2. kubectl port-forward 기반 접속

포트 포워딩은 쿠버네티스 내부의 서비스를 현재 여러분이 명령어를 입력하고 있는 머신의 포트와 직접 뚫어주는(연결해 주는) 기능입니다.

```bash
# web-frontend 서비스의 80번 포트를 8080 포트로 연결
# --address 0.0.0.0 옵션: 외부 네트워크(윈도우 브라우저, 다른 PC 등)에서의 접속을 허용

kubectl port-forward svc/web-frontend 8080:80 --address 0.0.0.0
```

위 명령어를 실행하면 터미널이 대기 상태가 되며 포워딩이 유지됩니다. (종료하려면 Ctrl + C를 누르세요.)

이제 웹 브라우저를 열고 자신의 실습 환경에 맞는 주소로 접속해 봅니다.

WSL 환경에서 실습 중인 경우 (Windows 브라우저에서 접속):
 - 주소창에 `http://localhost:8080` 입력
 - (--address 0.0.0.0 옵션 덕분에 WSL 내부의 포트가 Windows로 잘 전달됩니다.)
 - 주의: 원격 서버의 방화벽(UFW 등)이나 클라우드 보안 그룹에서 8080 포트가 열려있어야 합니다.

이 방식은 NodePort를 직접 쓰지 않고 로컬 포트로 간단히 확인할 때 유용합니다.


#### 14. 실습 종료 및 리소스 정리 (Cleanup)

실습이 끝난 후, 클러스터 자원을 낭비하지 않도록 배포했던 리소스를 삭제합니다. 생성할 때 apply를 썼다면, 지울 때는 delete를 사용합니다.

```Bash
# 배포했던 모든 파드와 서비스 일괄 삭제
kubectl delete -f k8s/manifests/

# K8s 리소스가 모두 지워졌는지 확인
kubectl get pods
kubectl get svc

# (선택) Minikube 클러스터 자체를 일시 정지하려면
minikube stop
```

#### WSL / VS Code 환경에서의 접속 이해

- `localhost`는 **현재 명령을 실행한 환경 기준**의 주소입니다.
- VS Code Remote 환경에서는:
  - 원격 머신의 포트가
  - 로컬 Windows의 `localhost`로 전달되어 보일 수 있습니다.
- 주소가 `localhost`로 보이더라도,
  실제 웹 서버는 **쿠버네티스 Service**입니다.

### 문제 1. Docker Compose와 Kubernetes의 차이점 
직전에 배운 Docker Compose와 Kubernetes의 가장 큰 차이점은 무엇인가요? (작동 환경을 중심으로 설명해 보세요.)

<details> <summary>정답 보기</summary>

Docker Compose는 '한 대의 컴퓨터(단일 서버)'에서 여러 컨테이너를 실행하고 묶어주는 도구인 반면, Kubernetes는 '여러 대의 컴퓨터(클러스터)'를 엮어서 다수의 컨테이너를 대규모로 자동 관리하고 오케스트레이션하는 도구입니다.

</details>


### 문제 2. 자기 치유 (Self-healing) 
새벽에 서버 한 대가 갑자기 고장 나서 컨테이너가 멈췄습니다. 이때 Kubernetes의 '자기 치유' 기능은 어떻게 동작하나요?

<details> <summary>정답 보기</summary>

장애를 즉시 감지하고, 죽어버린 컨테이너들을 다른 정상적인 서버로 자동으로 옮겨서 다시 실행시킵니다. 이를 통해 서비스 중단을 막고 가용성을 유지합니다.

</details>

### 문제 3. 자동 확장 (Auto-scaling) 
쇼핑몰 특가 이벤트로 평소보다 웹 사이트 접속자가 100배 폭주할 때, Kubernetes는 어떻게 대처하나요?

<details> <summary>정답 보기</summary>

트래픽 부하를 감지하여 자동으로 더 많은 컨테이너를 배포(확장)해 부하를 분산시킵니다. 이후 손님이 빠져나가면 다시 컨테이너 수를 줄여 자원을 최적화합니다.

</details>

여러 대의 배달 로봇(컨테이너)을 중앙에서 조율하며 고장 난 로봇은 즉시 교체하고, 주문이 몰리면 예비 로봇을 더 투입하는 "전국망 중앙 AI 통제 시스템"이나 "최고 책임 관리자"와 같습니다.

### 문제 7. 포트 포워딩 (Port-forwarding)을 사용하는 이유 
WSL이나 원격 서버 환경에서 실습할 때, 웹 서비스에 접속하기 위해 `minikube service` 명령어가 아닌 `kubectl port-forward`를 권장하는 이유는 무엇인가요?

<details> <summary>정답 보기</summary>

`minikube service`가 출력하는 주소(예: 192.168.x.x)는 Docker 컨테이너 내부의 가상 IP라 외부 브라우저에서 직접 접속할 수 없기 때문입니다. 포트 포워딩을 쓰면 쿠버네티스 내부와 내 PC의 포트를 직접 연결하여 안정적인 접속이 가능합니다.

</details>
