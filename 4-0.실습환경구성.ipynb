# Docker 설치

## 1. Windows - Docker Desktop

### 1.1 다운로드 링크
- [공식 Docker Desktop 다운로드](https://www.docker.com/products/docker-desktop/)

> 위 링크에서 **Windows용 Docker Desktop Installer**를 다운로드한다.





### 1.2 어떤 버전을 다운로드해야 하나?

#### ✅ amd64 (x86_64)
- Intel 또는 AMD CPU 기반 PC  
- 대부분의 Windows 데스크탑/노트북  
- 예: Intel i5, i7, i9 / AMD Ryzen 시리즈  

#### ✅ arm64
- ARM 기반 CPU 사용 시 선택  
- 예:  
  - Windows on ARM 기기 (예: Surface Pro X)  
  - 일부 ARM 서버 보드 (라즈베리파이 등)  

> ✔️ 대부분의 Windows 사용자는 **amd64** 버전을 설치하면 된다.





### 1.3 내 PC 아키텍처 확인 방법

#### (1) 시스템 정보 확인
1. **Win + Pause/Break** 키 → "시스템" 창 열기  
2. **시스템 종류** 확인  
   - "x64 기반 프로세서" → amd64 다운로드  
   - "ARM 기반 프로세서" → arm64 다운로드  

#### (2) PowerShell 명령어
```powershell
wmic cpu get caption
```

출력 예시:  
- Intel(R) Core(TM) i7-9750H → amd64  
- Qualcomm Snapdragon → arm64  





## 2. Linux - Docker Engine 직접 설치
리눅스에서는 Docker Desktop 대신 **Docker Engine**을 직접 설치한다.  
배포판별 설치 절차는 공식 문서를 따른다.

### 2.1 Ubuntu / Debian 계열
```bash
sudo apt-get update
sudo apt-get install -y docker.io
sudo systemctl enable --now docker
```

### 2.2 CentOS / RHEL 계열
```bash
sudo yum install -y docker
sudo systemctl enable --now docker
```

### 2.3 설치 확인
```bash
docker --version
docker run hello-world
```
