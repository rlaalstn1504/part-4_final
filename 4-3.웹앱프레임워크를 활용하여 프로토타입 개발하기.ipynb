### 0.5 Streamlit이란?

Streamlit은 **Python 코드만으로 데이터 앱을 빠르게 만들 수 있는 오픈소스 웹앱 프레임워크**입니다.  
프런트엔드 지식 없이도 입력·출력·시각화가 가능한 UI를 구성할 수 있어,  
**AI 모델 데모, 대시보드, 프로토타입** 제작에 특히 적합합니다.  

- **빠른 개발**: Python 스크립트를 곧바로 웹앱으로 실행
- **즉시 피드백**: 코드 수정 시 자동 리로드
- **시각화 통합**: Matplotlib/Plotly/Altair 등과 자연스럽게 연동
- **쉬운 배포**: Streamlit Cloud로 간단 배포 가능

이제 간단한 Hello 앱을 실행해 기본 동작을 확인해보겠습니다.


### 0.6 Streamlit Hello 실행

Streamlit은 명령어 한 줄로 웹 서버를 실행할 수 있는 Python 기반 프레임워크입니다.  
기본 동작을 익히기 위해 **Hello World 앱**을 실행해보며, Streamlit의 구조와 자동 리로드 기능을 확인합니다.

#### Streamlit 실행 테스트

1. `app.py` 파일을 생성하고 다음 내용을 입력합니다.

```python
import streamlit as st

st.title("Hello Streamlit 👋")
st.write("Streamlit 앱이 정상적으로 실행되고 있습니다.")
```

2. **터미널**에서 실행합니다.

```bash

```

3. 브라우저가 자동으로 열리며 `localhost:8501`에서 앱이 실행되는지 확인합니다.

4. 코드 수정 후 저장하면 브라우저가 자동으로 새로고침되는지 확인합니다.

## 1. AI 모델 배포 기본 개념

AI 모델 배포란, **학습된 모델을 실제 사용자에게 서비스 형태로 제공하는 과정**입니다.  
즉, 모델이 단순한 “코드”에서 **실제 운영 가능한 예측 시스템**으로 전환되는 단계입니다.


### 1.1 모델 배포의 개념과 주요 단계

AI 모델 배포(Deployment)란, **학습된 모델을 실제 사용자에게 서비스 형태로 제공하는 과정**입니다.  
즉, 모델이 단순한 코드에서 **운영 가능한 예측 시스템**으로 전환되는 단계입니다.

| 단계 | 설명 |
|------|------|
| 1️⃣ **모델 학습(Training)** | 오프라인 환경에서 데이터 기반 학습 수행 |
| 2️⃣ **모델 저장(Save/Export)** | `.pkl`, `.pt`, `.onnx` 등으로 모델 저장 |
| 3️⃣ **모델 로드(Load)** | 애플리케이션에서 모델 불러오기 |
| 4️⃣ **예측 서비스 구현** | FastAPI / Streamlit 등으로 예측 결과 반환 |
| 5️⃣ **배포(Deployment)** | 웹/클라우드 환경에서 서비스 공개 및 유지보수 |

**요약**  
> 모델 배포는 “학습된 모델을 코드로 통합하고(UI 연결),  
> 이를 클라우드 환경에서 실행 가능한 형태로 공개하는 과정”입니다.

### 1.2 웹앱 프레임워크의 역할 및 선택 기준

AI 모델을 웹 서비스로 제공하려면, 모델을 감싸는 **웹 프레임워크(Web Framework)** 가 필요하다.  
이는 사용자의 입력을 받아 모델의 예측 결과를 시각적으로 보여주는 역할을 합니다.

| 구분 | Streamlit | Gradio | FastAPI / Flask |
|------|------------|---------|-----------------|
| 주 목적 | 시각화 중심 앱 개발 | 모델 체험 데모 | API 서버 구축 |
| 특징 | Python 통합형, 즉시 실행 | 함수 기반 UI | REST 구조, 프론트엔드 분리 |
| 권장 활용 | 내부 프로토타입, 대시보드 | AI 데모 | 상용 서비스 |

**요약**  
- Streamlit → 빠른 프로토타입, 대시보드
- Gradio → 모델 체험용 데모  
- FastAPI/Flask → 확장형 백엔드 API

## 2. Streamlit 기본 구조 및 구성요소

이 장에서는 Streamlit의 **앱 실행 구조와 주요 컴포넌트 구성 방식**을 학습합니다.  
입력 위젯, 출력 요소, 레이아웃 구성, 그리고 `st.session_state`를 이용한 상태 관리 방법을 실습을 통해 익힌다.  
결과적으로 “**컴포넌트 카탈로그**”를 직접 만들어, 이후 프로젝트에서 재사용할 수 있는 UI 샘플 라이브러리를 구축합니다.

#### 기본 실행 원리
Streamlit 앱은 하나의 Python 스크립트를 **웹서버로 실행**하는 구조입니다.

1. `streamlit run app.py` 명령 실행  
2. Streamlit 런타임이 코드를 상단부터 실행  
3. 출력 결과를 **브라우저에 즉시 렌더링**

> Streamlit은 명령형 구조로, **코드를 위에서 아래로 실행하며 매번 새로 렌더링**합니다.

### 2.1 입력 위젯 (Input Widgets)

사용자 입력을 받는 UI 요소들입니다.  
모든 입력 위젯은 Streamlit 내부적으로 **자동 상태 저장(stateful)** 구조를 가진다.

| 위젯 | 설명 | 예시 코드 |
|------|------|-----------|
| `st.text_input()` | 문자열 입력 | `name = st.text_input("이름 입력")` |
| `st.number_input()` | 숫자 입력 | `age = st.number_input("나이", 0, 120)` |
| `st.slider()` | 범위 선택 | `level = st.slider("난이도", 1, 10, 5)` |
| `st.selectbox()` | 드롭다운 선택 | `lang = st.selectbox("언어", ["Python", "R", "C++"])` |
| `st.checkbox()` | 체크박스 | `agree = st.checkbox("동의합니다")` |
| `st.file_uploader()` | 파일 업로드 | `file = st.file_uploader("CSV 업로드")` |

#### 예시

```python
import streamlit as st

st.header("입력 위젯 데모")

name = st.text_input("이름을 입력하세요")
age = st.number_input("나이", min_value=0, max_value=120, value=25)
lang = st.selectbox("언어 선택", ["Python", "R", "C++"])
submit = st.button("확인")

if submit:
    st.success(f"{name}님은 {age}세이며, {lang}을(를) 사용합니다.")
```

### 2.2 출력 요소 (Output Components)

Streamlit은 데이터를 텍스트, 표, 차트 등 다양한 형태로 표현할 수 있다.

| 출력 함수 | 용도 | 예시 |
|------------|------|------|
| `st.write()` | 텍스트, 변수, 데이터프레임 자동 처리 | `st.write(df)` |
| `st.dataframe()` | 스크롤 가능한 표 | `st.dataframe(df)` |
| `st.table()` | 정적 표 | `st.table(df.head())` |
| `st.json()` | JSON 포맷 출력 | `st.json({"name": "AI", "ver": 1.0})` |
| `st.line_chart()` | 선형 차트 | `st.line_chart(df)` |
| `st.bar_chart()` | 막대 차트 | `st.bar_chart(df)` |

#### 예시

````python
import streamlit as st
import pandas as pd
import numpy as np

st.header("출력 요소 예제")

data = pd.DataFrame({
    "x": np.arange(1, 6),
    "y": np.random.randint(1, 100, 5)
})

st.dataframe(data)
st.bar_chart(data.set_index("x"))
````

### 2.3 레이아웃 구성 및 상태 관리

##### (1) 컬럼(Column) 레이아웃

st.columns()는 화면을 가로로 나누어 여러 영역을 병렬로 배치할 때 사용합니다.  
입력 영역과 결과 영역을 좌우로 분리할 때 유용합니다.

```python
col1, col2 = st.columns(2)

with col1:
    st.subheader("왼쪽 영역")
    st.write("입력 폼 또는 설명")

with col2:
    st.subheader("오른쪽 영역")
    st.write("그래프 또는 결과")
```

##### (2) 사이드바(Sidebar)

사이드바는 화면 왼쪽에 고정된 영역으로, 페이지 이동, 옵션 선택, 전역 설정 등에 주로 사용됩니다.

```python
st.sidebar.title("설정 메뉴")
option = st.sidebar.radio("페이지 선택", ["홈", "예측", "설정"])
st.write("현재 선택:", option)
```

##### (3) 컨테이너(Container)

컨테이너는 여러 UI 요소를 하나의 논리적 블록으로 묶는 용도로 사용됩니다.  
조건부 렌더링이나 코드 가독성 개선에 도움이 됩니다.

```python
container = st.container()
container.write("이곳은 별도 묶음 영역입니다.")
```


##### (4) 상태 관리 (`st.session_state`)

Streamlit은 매 실행마다 전체 스크립트를 다시 실행하지만,  
`st.session_state`를 이용하면 **입력값·상태를 유지**할 수 있습니다.  
세션(Session) 이란, 사용자가 앱에 접속해 있는 동안 유지되는 개별 사용자별 실행 상태 공간을 의미합니다.

```python
import streamlit as st

if "count" not in st.session_state:
    st.session_state.count = 0

if st.button("증가"):
    st.session_state.count += 1

st.write("현재 카운트:", st.session_state.count)
```

> **Tip:**  `st.session_state`는 페이지 전환이나 폼 리셋 시에도 데이터를 유지할 수 있어  
> UX 향상에 매우 유용하다.

Streamlit 앱은 기본적으로 **매 실행마다 전체 스크립트를 재실행**합니다.

```python
name = st.text_input("이름 입력") 
``` 

**아래 코드로 대체합니다**
```python
if "user" not in st.session_state:
    st.session_state.user = ""

name = st.text_input("이름 입력", st.session_state.user)
st.session_state.user = name
```

아래 코드를 `app.py`의 상단에 추가해서 확인해보세요.

```python
if "user" in st.session_state and st.session_state.user:
```

## 5. Streamlit을 활용한 AI 모델 통합

이 장에서는 Streamlit 앱에 **AI 모델을 직접 연동**하여  
사용자 입력 → 예측 수행 → 결과 시각화까지 이어지는 **단방향 파이프라인**을 완성합니다.  

모델 로딩, 예측, 예외 처리, 데이터 시각화, 로그 저장까지 모두 실습 중심으로 다룬다.

### 5.1 모델 로딩 및 예측 함수 구현

Streamlit은 매번 코드가 재실행되기 때문에,  
모델을 효율적으로 로드하기 위해 **캐싱 데코레이터(`@st.cache_resource`)** 를 사용합니다.  
이를 통해 모델이 매번 새로 학습되거나 로드되는 문제를 방지합니다.
해당 내용은 뒤에서 자세히 다루겠습니다.

###  `src/model.py` 구현 예시

```python
# src/model.py
import streamlit as st
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

@st.cache_resource
def load_model():
    """더미 텍스트 분류 모델 로드"""
    X = ["good", "great", "awesome", "bad", "awful", "terrible"]
    y = [1, 1, 1, 0, 0, 0]
    model = Pipeline([
        ("vec", CountVectorizer()),
        ("clf", LogisticRegression(max_iter=200))
    ])
    model.fit(X, y)
    return model

@st.cache_data
def predict_texts(texts):
    """입력 텍스트 리스트 예측"""
    model = load_model()
    preds = model.predict(texts)
    probs = model.predict_proba(texts)[:, 1]
    df = pd.DataFrame({"text": texts, "label": preds, "prob": probs})
    return df
```

> 💡 `@st.cache_resource`는 모델 객체처럼 **리소스(무거운 객체)** 에 사용  
> `@st.cache_data`는 데이터 연산 결과 캐싱에 사용

```python
import streamlit as st
import pandas as pd

LOG_PATH = "data/log.csv"

def load_log():
    """로그 파일 불러오기"""
    try:
        df = pd.read_csv(LOG_PATH, names=["text", "label", "prob", "timestamp"])
        df["timestamp"] = pd.to_datetime(df["timestamp"])

        return pd.DataFrame(columns=["text", "label", "prob", "timestamp"])
```

```python
import streamlit as st
from src.model import predict_texts
import pandas as pd
import time

st.header("🔮 텍스트 예측")             # 페이지 상단 헤더 출력

text = st.text_input("문장 입력", "good")  # 사용자로부터 예측할 문장을 입력받음 (기본값: "good")

if st.button("예측 실행"):              # 사용자가 버튼을 클릭했을 때만 아래 코드 실행
    t0 = time.perf_counter()           # 예측 시작 시점의 고해상도 시간 기록
    df = predict_texts([text])         # 입력 문장을 리스트로 감싸 모델에 전달하여 예측 수행
    dt = time.perf_counter() - t0      # 예측 처리에 걸린 총 시간 계산

    label = int(df["label"][0])        # 예측 결과 라벨을 정수형으로 추출
    prob = float(df["prob"][0])         # 예측 확률 값을 실수형으로 추출

    st.metric(label="예측 결과", value=f"{label}", delta=f"{prob*100:.1f}% 확신")  # 예측 라벨과 확신도(%)를 지표 형태로 표시
    st.toast("예측 완료! 🎉")           # 예측이 끝났음을 알리는 토스트 알림 표시
    st.info(f"처리 시간: {dt:.3f}s")    # 예측 처리 시간을 정보 메시지로 출력

    df["timestamp"] = pd.Timestamp.now()  # 예측 시점을 타임스탬프로 데이터프레임에 추가
    df.to_csv("data/log.csv", mode="a", header=False, index=False)  # 예측 결과를 로그 CSV 파일에 누적 저장
```

## 8. 성능 최적화 및 실행 효율화 → 전/후 비교 실험

이 장에서는 **Streamlit 애플리케이션의 성능을 최적화**하는 방법을 학습합니다.  
`@st.cache_data`, `@st.cache_resource`를 활용해 **모델 로드·데이터 처리 속도 개선**을 수치화하고,  
**메모리 관리·배포 환경 차이 대응 전략**까지 실습합니다.

###  8.1 Streamlit 캐싱의 개념

Streamlit 앱은 위젯 변경, 브라우저 새로고침(F5), 사용자 접속 과 같은 이벤트가 발생할 때마다  
**스크립트를 위에서부터 끝까지 다시 실행(rerun)** 합니다.

이때, 대용량 데이터나 모델을 매번 다시 불러오면 앱 속도가 크게 느려질 수 있습니다.
이를 해결하기 위해 Streamlit은 목적이 다른 두 가지 캐시 데코레이터를 제공합니다.

| 구분 | 캐시 대상 | 반환 방식 | 특징 |
|------|-----------|-----------|------|
| `@st.cache_data` | 함수의 **결과값** | 매번 **새 복사본** 반환 | 안전 (변이 영향 없음) |
| `@st.cache_resource` | **객체 인스턴스 자체** | 항상 **같은 객체** 반환 | 효율적 (변이 주의) |

- Streamlit은 **항상 rerun 된다**
- 캐시는 rerun을 막는 게 아니라 **비싼 연산을 다시 하지 않게 해주는 장치**
- `cache_data`는 결과 중심 캐시
- `cache_resource`는 객체 공유 캐시

### 선택 기준 한 줄 요약
가공·수정할 결과 → `cache_data`  
같은 인스턴스를 계속 써야 하는 객체 → `cache_resource`

### 8.2 캐싱으로 모델/데이터 로드 속도 향상

####  예시 ① 데이터 캐싱
```python
import streamlit as st
import pandas as pd
import time

@st.cache_data
def load_large_csv():
    time.sleep(3)  # 로드 지연 시뮬레이션
    df = pd.DataFrame({"A": range(1000), "B": range(1000, 2000)})
    return df

st.header("📁 캐싱 데이터 로드 테스트")

t0 = time.perf_counter()
df = load_large_csv()
dt = time.perf_counter() - t0

st.write(f"로드 시간: {dt:.2f}초")
st.dataframe(df.head())
```

- 첫 실행: 약 3초 지연  
- 두 번째 실행: 즉시 반환 (캐시 사용)

#### 예시 ② 모델 리소스 캐싱
```python
import streamlit as st
import time
from sklearn.linear_model import LogisticRegression
import numpy as np

@st.cache_resource
def load_model():
    time.sleep(2)
    model = LogisticRegression()
    X = np.array([[1, 0], [0, 1]])
    y = [1, 0]
    model.fit(X, y)
    return model

st.header("🤖 캐시된 모델 로드")

t0 = time.perf_counter()
model = load_model()
dt = time.perf_counter() - t0

st.success(f"모델 로드 완료 ({dt:.2f}s)")
```

> 💡 **결과 해석**  
> - 첫 실행: 학습 포함 약 2초 소요  
> - 이후 실행: 0.00~0.02초 내 즉시 반환  



### 8.3 Before / After 성능 비교 실험

####  실습 코드 — 성능 측정 및 시각화

```python
import streamlit as st
import time
import pandas as pd
import altair as alt

@st.cache_data
def slow_function(n):
    time.sleep(n)
    return n

st.header("⚡ 캐시 전/후 성능 비교")

results = []
for n in [1, 1, 1]:
    t0 = time.perf_counter()
    slow_function(1)
    dt = time.perf_counter() - t0
    results.append(dt)

df = pd.DataFrame({"run": [1, 2, 3], "time": results})

st.table(df)

chart = alt.Chart(df).mark_line(point=True).encode(
    x="run:O", y="time:Q"
).properties(title="캐싱 전/후 시간 변화 (초)")
st.altair_chart(chart, use_container_width=True)
```

> **결과**:  
> - 첫 번째 실행 약 1초, 이후 0.00~0.01초  
> - 속도 개선률 약 **98~99%**



### 8.4 이미지·모델 파일 경량화 전략

| 항목 | 방법 | 설명 |
|------|------|------|
| **이미지** | JPEG/WEBP 포맷, 해상도 축소 | Cloud 메모리 절약 |
| **모델 파일** | Pickle → Gzip 압축 또는 `.pkl.gz` | 파일 크기 70% 감소 |
| **모델 경량화** | HuggingFace Hub 모델 링크 사용 | 로컬 업로드 불필요 |
| **데이터** | CSV 대신 Feather/Parquet 사용 | I/O 속도 향상 (10~50배) |

> ⚡ **예시: Pickle 압축 저장**
```python
import pickle, gzip

with gzip.open("model.pkl.gz", "wb") as f:
    pickle.dump(model, f)
```
